# 1. Partial Derivatives

Misalkan suatu fungsi bergantung pada beberapa variabel

$$
u = u(x,y)
$$

Persamaan ini menyatakan bahwa nilai fungsi (u) ditentukan oleh dua variabel bebas yaitu (x) dan (y). Jika salah satu variabel berubah maka nilai fungsi juga dapat berubah.


Turunan parsial menyatakan perubahan fungsi terhadap **satu variabel**, dengan variabel lainnya dianggap konstan.

Artinya ketika kita menghitung turunan parsial terhadap satu variabel, variabel lain dianggap tetap atau tidak berubah.

Turunan parsial terhadap $x$:

$$
\frac{\partial u}{\partial x}
=

\lim_{h\to0}
\frac{u(x+h,y)-u(x,y)}{h}
$$

Persamaan ini adalah definisi limit dari turunan parsial terhadap (x). Kita melihat perubahan nilai fungsi ketika (x) berubah sedikit sebesar (h), sementara (y) tetap.

dan terhadap $y$:

$$
\frac{\partial u}{\partial y}
=

\lim_{h\to0}
\frac{u(x,y+h)-u(x,y)}{h}
$$

Definisi ini menunjukkan perubahan fungsi ketika variabel (y) berubah sedikit, sedangkan (x) dianggap konstan.

Notasi yang sering digunakan adalah

$$
u_x=\frac{\partial u}{\partial x}
$$

Notasi (u_x) adalah bentuk singkat dari turunan parsial terhadap (x).

$$
u_y=\frac{\partial u}{\partial y}
$$

Notasi (u_y) merupakan bentuk singkat dari turunan parsial terhadap (y).

# Second Order Partial Derivatives

Turunan parsial dapat diambil lebih dari satu kali.

Artinya setelah memperoleh turunan pertama, kita masih dapat menurunkannya lagi untuk memperoleh turunan orde yang lebih tinggi.
Turunan kedua terhadap $x$:

$$
u_{xx}=
\frac{\partial}{\partial x}
\left(
\frac{\partial u}{\partial x}
\right)
$$

Ini berarti kita mengambil turunan terhadap (x) dari turunan pertama terhadap (x).

$$
\boxed{
u_{xx}=\frac{\partial^2 u}{\partial x^2}
}
$$

Notasi ini adalah bentuk ringkas dari turunan parsial kedua terhadap variabel (x).

Turunan kedua terhadap $y$:

$$
\boxed{
u_{yy}=\frac{\partial^2 u}{\partial y^2}
}
$$

Ini menyatakan turunan parsial kedua terhadap variabel (y).

## Mixed Derivatives

Turunan parsial juga dapat diambil secara campuran

Artinya turunan dilakukan terhadap dua variabel yang berbeda secara berurutan.

$$
u_{xy}
=

\frac{\partial}{\partial x}
\left(
\frac{\partial u}{\partial y}
\right)
$$

Ini berarti kita menurunkan fungsi terhadap (y) terlebih dahulu, kemudian hasilnya diturunkan lagi terhadap (x).

$$
\boxed{
u_{xy}=
\frac{\partial^2 u}{\partial x \partial y}
}
$$

Notasi ini menyatakan turunan parsial campuran terhadap (y) lalu (x).

atau

$$
u_{yx}
=

\frac{\partial}{\partial y}
\left(
\frac{\partial u}{\partial x}
\right)
$$

Pada bentuk ini urutan turunan dibalik, yaitu pertama terhadap (x) kemudian terhadap (y).

## Equality of Mixed Partial Derivatives

Pertanyaan penting dalam kalkulus multivariabel:

apakah

$$
u_{xy}=u_{yx}
$$

selalu berlaku?

Pertanyaan ini menanyakan apakah hasil turunan campuran akan sama meskipun urutan diferensiasinya ditukar.


Jawabannya **tidak selalu**, tetapi berlaku jika fungsi cukup halus.

Jika fungsi memiliki turunan parsial yang kontinu hingga orde tertentu, maka kedua turunan campuran tersebut akan sama.


## Teorema Clairaut (Schwarz)

Jika fungsi $u(x,y)$ memiliki turunan parsial kontinu hingga orde dua pada suatu domain, maka

$$
\boxed{
u_{xy}=u_{yx}
}
$$

Teorema ini menyatakan bahwa jika semua turunan parsial orde dua kontinu maka urutan diferensiasi dapat dipertukarkan.

## Penjelasan Matematis

Misalkan

$$
u_x = \frac{\partial u}{\partial x}
$$

Ini adalah turunan pertama fungsi terhadap variabel (x).

Turunan terhadap $y$:

$$
u_{xy} =
\frac{\partial}{\partial y}
\left(
\frac{\partial u}{\partial x}
\right)
$$

Ini menunjukkan bahwa turunan pertama terhadap (x) kemudian diturunkan lagi terhadap (y).

Sebaliknya

$$
u_{yx} =
\frac{\partial}{\partial x}
\left(
\frac{\partial u}{\partial y}
\right)
$$

Pada bentuk ini turunan pertama terhadap (y) kemudian diturunkan lagi terhadap (x).


Jika turunan kedua kontinu maka limit pertukaran diferensiasi dapat ditukar sehingga

$$
\boxed{
\frac{\partial^2 u}{\partial x \partial y}
=

\frac{\partial^2 u}{\partial y \partial x}
}
$$

Kontinuitas turunan kedua memungkinkan operasi diferensiasi dipertukarkan tanpa mengubah hasilnya.

# 2. Classification of Partial Differential Equations

Bentuk umum **PDE linear orde dua dengan dua variabel** adalah

$$
a u_{xx}+2b u_{xy}+c u_{yy}+d u_x+e u_y+fu=g
$$

Persamaan ini merupakan bentuk umum dari persamaan diferensial parsial linear orde dua dengan dua variabel bebas.

dengan

$$
a,b,c,d,e,f,g
$$

merupakan fungsi dari $x$ dan $y$.

Artinya koefisien-koefisien dalam persamaan dapat bergantung pada posisi dalam domain.

Untuk klasifikasi PDE, kita hanya memperhatikan **bagian orde dua**

$$
a u_{xx}+2b u_{xy}+c u_{yy}
$$

Klasifikasi jenis PDE ditentukan hanya oleh koefisien turunan orde dua.

## Hubungan dengan Bentuk Kuadrat

Ekspresi tersebut analog dengan bentuk kuadrat

$$
a(dx)^2 +2b(dx)(dy)+c(dy)^2
$$
yang identik dengan persamaan kerucut

$$
Ax^2+Bxy+Cy^2
$$

Persamaan ini digunakan untuk menggambarkan kurva kerucut seperti elips, parabola, dan hiperbola.

Diskriminan dari bentuk kuadrat tersebut adalah

$$
\Delta = B^2-4AC
$$

Diskriminan digunakan untuk menentukan jenis kurva yang dihasilkan.

Pada PDE koefisiennya adalah

$$
B=2b
$$

Koefisien (B) pada bentuk kuadrat ekuivalen dengan (2b) pada persamaan PDE.


sehingga

$$
\Delta=(2b)^2-4ac
$$

Diskriminan dinyatakan dalam bentuk koefisien PDE.
$$
\Delta=4(b^2-ac)
$$

Bentuk ini diperoleh dengan menyederhanakan persamaan diskriminan.
Karena faktor 4 tidak mempengaruhi tanda diskriminan, maka klasifikasi hanya bergantung pada

$$
\boxed{b^2-ac}
$$

Nilai tanda dari ekspresi ini menentukan tipe PDE.

## Tipe Persamaan

### Elliptic Equation

Jika

$$
\boxed{b^2-ac<0}
$$

maka PDE disebut **elliptic**.

Persamaan elliptic biasanya menggambarkan fenomena keadaan tunak seperti distribusi temperatur pada keadaan stabil.

Contoh klasik:

$$
\boxed{
u_{xx}+u_{yy}=0
}
$$

yang dikenal sebagai **Laplace equation**.

Persamaan Laplace sering muncul dalam fisika seperti potensial listrik dan aliran panas pada keadaan steady state.

### Parabolic Equation

Jika

$$
\boxed{b^2-ac=0}
$$

maka disebut **parabolic**.

Jenis persamaan ini biasanya menggambarkan proses difusi atau penyebaran.

Contoh:

$$
\boxed{
u_t=u_{xx}
}
$$

yaitu **heat equation**.

Persamaan panas menggambarkan bagaimana temperatur menyebar di dalam suatu medium seiring waktu.

### Hyperbolic Equation

Jika

$$
\boxed{b^2-ac>0}
$$

maka disebut **hyperbolic**.

Persamaan ini biasanya menggambarkan fenomena propagasi gelombang.

Contoh:

$$
\boxed{
u_{tt}-u_{xx}=0
}
$$

yaitu **wave equation**.

Persamaan gelombang menggambarkan perambatan gelombang seperti gelombang suara atau gelombang pada tali.

# 3. Fourier Analysis

Fourier analysis merupakan alat penting dalam penyelesaian PDE karena memungkinkan kita mengekspresikan fungsi sebagai kombinasi sinus dan cosinus.

Metode ini sangat berguna dalam menyelesaikan PDE dengan kondisi batas tertentu.

Misalkan fungsi periodik $f(x)$ pada interval

$$
[-L,L]
$$

Artinya fungsi tersebut berulang secara periodik pada interval tersebut.

maka fungsi tersebut dapat direpresentasikan sebagai

$$
f(x)=a_0+
\sum_{n=1}^{\infty}
a_n\cos\left(\frac{n\pi x}{L}\right)
+
\sum_{n=1}^{\infty}
b_n\sin\left(\frac{n\pi x}{L}\right)
$$

Persamaan ini disebut deret Fourier yang menyatakan fungsi sebagai kombinasi tak hingga dari fungsi sinus dan cosinus.

Koefisien Fourier diberikan oleh

$$
\boxed{
a_n=
\frac{1}{L}
\int_{-L}^{L}
f(x)
\cos\left(\frac{n\pi x}{L}\right)dx
}
$$

Koefisien (a_n) menentukan kontribusi komponen cosinus dalam deret Fourier.

$$
\boxed{
b_n=
\frac{1}{L}
\int_{-L}^{L}
f(x)
\sin\left(\frac{n\pi x}{L}\right)dx
}
$$

Koefisien (b_n) menentukan kontribusi komponen sinus dalam deret Fourier.

Representasi ini sangat penting dalam penyelesaian PDE menggunakan **separation of variables**.

Metode pemisahan variabel sering menghasilkan solusi dalam bentuk deret Fourier.

## Fourier Transform

Untuk fungsi pada domain tak terbatas, digunakan **Fourier transform**.

Transformasi ini mengubah fungsi dari domain ruang ke domain frekuensi.

Transformasi Fourier didefinisikan sebagai

$$
\boxed{
\hat v(\xi)
=

\frac{1}{\sqrt{2\pi}}
\int_{-\infty}^{\infty}
v(x)e^{-i\xi x}dx
}
$$

Transformasi ini mengubah fungsi (v(x)) menjadi fungsi frekuensi (\hat v(\xi)).

Transformasi baliknya adalah

$$
\boxed{
v(x)
=

\frac{1}{\sqrt{2\pi}}
\int_{-\infty}^{\infty}
\hat v(\xi)e^{i\xi x}d\xi
}
$$

Transformasi balik digunakan untuk mengembalikan fungsi dari domain frekuensi ke domain ruang.

## Fourier Transform of Derivatives

Salah satu alasan utama penggunaan Fourier transform dalam PDE adalah karena turunan berubah menjadi operasi aljabar.

Hal ini membuat persamaan diferensial menjadi lebih mudah diselesaikan.

Misalkan

$$
\widehat{v_x}
=

\frac{1}{\sqrt{2\pi}}
\int_{-\infty}^{\infty}
\frac{dv}{dx}e^{-i\xi x}dx
$$

Ini adalah transformasi Fourier dari turunan pertama fungsi (v(x)).

Gunakan integrasi parsial

Metode ini digunakan untuk menyederhanakan integral yang melibatkan turunan.
$$
u=e^{-i\xi x}
$$

Ini dipilih sebagai fungsi (u) dalam integrasi parsial.

$$
dv=v_x dx
$$

Ini dipilih sebagai bagian diferensial dalam integrasi parsial.

maka

$$
du=-i\xi e^{-i\xi x}dx
$$

Turunan dari fungsi eksponensial kompleks menghasilkan faktor (-i\xi).

$$
v=v(x)
$$

Integral dari (v_x) menghasilkan kembali fungsi (v(x)).

Sehingga

$$
\int v_x e^{-i\xi x}dx
=

v e^{-i\xi x}
\int v(-i\xi e^{-i\xi x})dx
$$

Hasil ini diperoleh dari penerapan rumus integrasi parsial.

Jika $v(x)\to0$ pada batas tak hingga maka suku pertama hilang sehingga diperoleh

Artinya nilai fungsi pada batas integral tak hingga dianggap mendekati nol.

$$
\widehat{v_x}
=

i\xi
\frac{1}{\sqrt{2\pi}}
\int v(x)e^{-i\xi x}dx
$$

Hasil ini menunjukkan bahwa transformasi Fourier dari turunan pertama menghasilkan faktor (i\xi).

$$
\boxed{
\widehat{v_x}=i\xi \hat v
}
$$

Artinya diferensiasi di ruang fisik menjadi perkalian oleh (i\xi) di ruang frekuensi.

Untuk turunan kedua

$$
\boxed{
\widehat{v_{xx}}=-\xi^2 \hat v
}
$$

Turunan kedua berubah menjadi perkalian dengan (-\xi^2) dalam domain frekuensi.


## Advection Equation

Pertimbangkan persamaan adveksi linear

$$
\boxed{
u_t+cu_x=0
}
$$

Persamaan ini menggambarkan transport suatu besaran dengan kecepatan konstan (c).

dengan kondisi awal

$$
u(x,0)=u_0(x)
$$

Ini menyatakan nilai awal fungsi pada waktu (t=0).

# Fourier Transform

Ambil transform terhadap $x$

$$
\widehat{u_t}+c\widehat{u_x}=0
$$

Transformasi Fourier diterapkan pada setiap suku persamaan.

Gunakan sifat transform

Sifat transform mempermudah perubahan turunan menjadi operasi aljabar.

$$
\widehat{u_t}=\partial_t\hat u
$$

Turunan terhadap waktu tetap menjadi turunan terhadap waktu setelah transformasi.

$$
\widehat{u_x}=i\xi \hat u
$$

Turunan terhadap ruang berubah menjadi perkalian dengan (i\xi).

Sehingga diperoleh

$$
\boxed{
\partial_t \hat u + ic\xi \hat u =0
}
$$

Persamaan PDE berubah menjadi persamaan diferensial biasa terhadap waktu.

**Penyelesaian ODE**

Persamaan tersebut merupakan ODE

$$
\frac{d\hat u}{dt}=-ic\xi\hat u
$$

Ini adalah persamaan diferensial linear orde satu.

Pisahkan variabel

$$
\frac{d\hat u}{\hat u}=-ic\xi dt
$$

Variabel dipisahkan agar dapat diintegrasikan.

Integrasi menghasilkan

$$
\ln \hat u = -ic\xi t +C
$$

Hasil integrasi memberikan bentuk logaritmik dari solusi.

Sehingga

$$
\hat u = Ce^{-ic\xi t}
$$

Solusi diperoleh dengan menghilangkan logaritma.

Dengan kondisi awal

$$
\hat u(\xi,0)=\hat u_0(\xi)
$$

Ini adalah transformasi Fourier dari kondisi awal.

maka

$$
\boxed{
\hat u(\xi,t)=\hat u_0(\xi)e^{-ic\xi t}
}
$$

Konstanta integrasi ditentukan dari kondisi awal.

**Inverse Transform**

$$
u(x,t)=
\frac{1}{\sqrt{2\pi}}
\int
\hat u_0(\xi)
e^{-ic\xi t}
e^{i\xi x}
d\xi
$$

Transformasi balik digunakan untuk kembali ke domain ruang.


Gabungkan eksponen

$$
e^{i\xi(x-ct)}
$$

Eksponen digabungkan menggunakan sifat eksponensial.


sehingga

$$
\boxed{
u(x,t)=u_0(x-ct)
}
$$

Solusi ini menunjukkan bahwa profil awal (u_0(x)) bergerak tanpa perubahan bentuk dengan kecepatan (c).


# 4. Elliptic Equations

Persamaan **elliptic** biasanya muncul pada **boundary value problem**, yaitu masalah dimana nilai solusi ditentukan oleh **kondisi batas domain**.

Contoh paling terkenal adalah **Laplace equation**

$$
\nabla^2 u = 0
$$

Persamaan ini adalah kasus khusus dari **Poisson equation**

$$
\nabla^2 u = f
$$

dimana

* $u$ = fungsi yang ingin dicari
* $f$ = sumber (source term)

Jika $f=0$ maka persamaan menjadi **Laplace equation**.

Persamaan elliptic biasanya menggambarkan **keadaan setimbang (steady state)** dari suatu sistem fisik, dimana tidak ada perubahan terhadap waktu dan distribusi hanya bergantung pada variabel ruang.

## Interpretasi Fisik

Persamaan ini sering muncul pada **steady-state system**.

Contoh:

* distribusi temperatur setelah waktu sangat lama
* potensial listrik tanpa muatan
* aliran fluida tak kental

Pada **steady state** sistem tidak berubah terhadap waktu, sehingga turunan waktu hilang dan yang tersisa hanya turunan ruang.

## Operator Linear Umum

Secara umum operator linear orde dua dapat ditulis

$$
L =
\sum_{i,j=1}^{n}
A_{ij}
\frac{\partial^2}{\partial x_i \partial x_j}
+
\sum_{i=1}^{n}
B_i
\frac{\partial}{\partial x_i}
+
C
$$

Operator ini disebut **elliptic** jika matriks koefisien orde dua

$$
A =
\begin{pmatrix}
A_{11} & A_{12} & \dots \
A_{21} & A_{22} & \dots \
\dots & \dots & \dots
\end{pmatrix}
$$

bersifat **positive definite** atau **negative definite**.

Artinya

$$
v^T A v
$$

memiliki tanda yang sama untuk semua

$$
v \neq 0, \qquad v \in \mathbb{R}^n
$$

Kondisi ini memastikan bahwa solusi dari PDE elliptic biasanya **halus (smooth)** dan dipengaruhi oleh seluruh kondisi batas pada domain.

# 5. Parabolic Equations

Persamaan **parabolic** biasanya menggambarkan **proses difusi atau konduksi**, dimana perubahan sistem terjadi secara bertahap terhadap waktu.

Pada persamaan ini, gangguan pada suatu titik akan **menyebar secara perlahan ke seluruh domain**, sehingga sistem cenderung menuju keadaan setimbang.

Bentuk umum

$$
u_t - Lu = f
$$

Contoh klasik adalah **heat equation**

$$
u_t=\nabla^2 u
$$

Persamaan ini memodelkan **penyebaran panas dalam suatu medium**, seperti batang logam yang dipanaskan.

### Heat Equation 1D

Pertimbangkan

$$
u_t=\kappa u_{xx}
$$

dengan kondisi

$$
u(x,0)=f(x)
$$

$$
u(0,t)=0
$$

$$
u(L,t)=0
$$

Parameter

$$
\kappa
$$

disebut **difusivitas termal**, yang menunjukkan seberapa cepat panas menyebar dalam material.

Jika $\kappa$ besar maka panas menyebar lebih cepat, sedangkan jika kecil maka proses difusi berlangsung lebih lambat.

## Separation of Variables

Misalkan

$$
u(x,t)=G(t)\phi(x)
$$

Turunan

$$
u_t=G'(t)\phi(x)
$$

$$
u_{xx}=G(t)\phi''(x)
$$

Substitusi

$$
G'\phi=\kappa G\phi''
$$

bagi dengan $G\phi$

$$
\frac{G'}{G}=\kappa\frac{\phi''}{\phi}
$$

Misalkan sama dengan $-\kappa\lambda$

sehingga

$$
G'=-\kappa\lambda G
$$

$$
\phi''+\lambda\phi=0
$$

## Solusi ODE Waktu

$$
G(t)=Ce^{-\kappa\lambda t}
$$

Faktor eksponensial ini menunjukkan bahwa **komponen solusi akan semakin mengecil terhadap waktu**, sehingga sistem secara alami menuju keadaan setimbang.

## Solusi ODE Ruang

$$
\phi(x)=A\cos(\sqrt{\lambda}x)+B\sin(\sqrt{\lambda}x)
$$

Gunakan boundary

$$
\phi(0)=0
$$

maka

$$
A=0
$$

dan

$$
\lambda=\left(\frac{n\pi}{L}\right)^2
$$

### Solusi Heat Equation

$$
u_n(x,t)=
B_n
e^{-\left(\frac{n\pi}{L}\right)^2\kappa t}
\sin\left(\frac{n\pi x}{L}\right)
$$

Superposisi

$$
u(x,t)=
\sum_{n=1}^{\infty}
B_n
e^{-\left(\frac{n\pi}{L}\right)^2\kappa t}
\sin\left(\frac{n\pi x}{L}\right)
$$

Koefisien

$$
B_n=
\frac{2}{L}
\int_0^L
f(x)
\sin\left(\frac{n\pi x}{L}\right)dx
$$

Solusi ini menunjukkan bahwa **bentuk awal temperatur akan menyebar dan semakin halus seiring waktu**.


# 6. Hyperbolic Equations

Persamaan **hyperbolic** sering muncul dalam bentuk **hukum konservasi (conservation laws)** atau **balance laws**, yaitu persamaan yang menyatakan bahwa suatu besaran fisik tidak hilang tetapi hanya berpindah atau berubah bentuk.

Contohnya adalah:

* hukum kekekalan momentum pada **Hukum II Newton**
* hukum kekekalan energi
* hukum kekekalan massa dalam dinamika fluida

Dalam banyak sistem fisik, perubahan suatu kuantitas dapat dinyatakan sebagai **perubahan terhadap waktu ditambah fluks terhadap ruang**. Bentuk matematis dari hukum konservasi ini sering menghasilkan **persamaan diferensial parsial tipe hiperbolik**.

Contoh klasik dari persamaan hiperbolik adalah **wave equation**

$$
u_{tt} - c^2 u_{xx} = 0
$$

Persamaan ini menggambarkan **propagasi gelombang** dalam suatu medium, misalnya:

* gelombang pada tali
* gelombang suara
* gelombang air
* gelombang elastik pada padatan

Parameter

$$
c
$$

menyatakan **kecepatan rambat gelombang** dalam medium.

Namun dalam analisis numerik, persamaan orde dua seperti wave equation sering diubah menjadi **sistem persamaan orde satu**.

Sebagai contoh

$$
u_t + A u_x = 0
$$

dimana

* $u$ adalah vektor variabel solusi
* $A$ adalah matriks koefisien

Sistem ini disebut **hyperbolic** jika matriks $A$ memiliki:

* **eigenvalue nyata (real eigenvalues)**
* **dapat didiagonalkan (diagonalizable)**

Kondisi ini memastikan bahwa solusi dapat dipropagasikan sepanjang **arah karakteristik tertentu**.

### Solving Hyperbolic Equations

Pendekatan paling umum untuk menyelesaikan persamaan hiperbolik adalah menggunakan **method of characteristics**.

Ide utama metode ini adalah mencari **kurva karakteristik** pada bidang

$$
(x,t)
$$

yang menunjukkan **jalur propagasi informasi dalam sistem**.

Sepanjang kurva ini, PDE dapat direduksi menjadi **persamaan diferensial biasa (ODE)** yang lebih mudah diselesaikan.

Jika persamaan bersifat **nonlinear**, karakteristik dapat:

* **berpotongan** → menghasilkan **shock**
* **menyebar** → menghasilkan **rarefaction wave**

Fenomena ini sangat penting dalam metode numerik karena **diskontinuitas (shock)** sering menyebabkan kesulitan dalam komputasi.

### Method of Characteristics

Metode karakteristik dimulai dengan asumsi bahwa solusi dapat dituliskan dalam bentuk

$$
u(x,t) = u(x(t), t)
$$

Artinya solusi mengikuti suatu kurva tertentu yang disebut **kurva karakteristik**.

Misalkan kita ingin menyelesaikan PDE

$$
u_t + c(x,t,u) u_x = 0
$$

Ambil turunan total terhadap waktu

$$
\frac{\text{d} u}{\text{d} t} =
\frac{\partial u}{\partial t}
+
\frac{\text{d} x}{\text{d} t}
\frac{\partial u}{\partial x}
$$

Setelah disubstitusi ke PDE, persamaan tersebut dapat diubah menjadi dua ODE

$$
\begin{aligned}
&\frac{\text{d} u}{\text{d} t} = 0, \quad \text{dan}\
&\frac{\text{d} x}{\text{d} t} = c(x,t,u).
\end{aligned}
$$

Persamaan pertama menunjukkan bahwa

$$
u(x(t),t)
$$

**konstan sepanjang kurva karakteristik**.

Dengan kata lain

$$
u(x,t) = u(x(t), 0)
$$

Sehingga solusi PDE dapat diperoleh dengan **menentukan bentuk kurva karakteristik $x(t)$**.

### Linear Advection Equation

Sebagai contoh sederhana, pertimbangkan persamaan

$$
\begin{aligned}
&u_t + c u_x = 0, \quad \Omega = (-\infty, \infty) \times (0, T) \
&u(x,0) = f(x), \quad \text{dan} \
&u_x(\pm \infty, t) \rightarrow 0.
\end{aligned}
$$

Persamaan karakteristiknya adalah

$$
\frac{\text{d} x}{\text{d} t} = c
$$

Solusinya

$$
x(t) = c t + x_0
$$

Artinya kurva karakteristik adalah **garis lurus** pada bidang $(x,t)$ dengan kemiringan $c$.

![Constant Wave Speed](./images/characteristics.png)

Persamaan garis karakteristik adalah

$$
x(t) = c t + x_0
$$

Solusi PDE kemudian menjadi

$$
u(x,t) = f(x - ct)
$$

yang berarti **profil awal hanya bergeser tanpa berubah bentuk**.

### Nonlinear Case: Burgers Equation

Untuk persamaan non-linear, kecepatan karakteristik bergantung pada solusi.

Contohnya adalah **Burgers equation**

$$
u_t + u u_x = 0
$$

Misalkan kondisi awal

$$
u(x,0) =
\begin{cases}
4, & x < 0, \\
3, & x > 0
\end{cases}
$$

Karakteristiknya adalah

$$
x(t) =
\begin{cases}
4t + x_0, & x < 0, \\
3t + x_0, & x > 0
\end{cases}
$$

Karena karakteristik dengan nilai 4 bergerak lebih cepat daripada yang bernilai 3, maka kurva akan **saling berpotongan**.

Hal ini menyebabkan terbentuknya **shock**.


![Shock](./images/characteristics_shock.png)

### Shock Speed (Rankine–Hugoniot)

Lokasi shock dapat dihitung menggunakan **Rankine–Hugoniot condition**

$$
\frac{\text{d} x_s}{\text{d} t} =
\frac{f(x_+, t) - f(x_-, t)}{u_+ - u_-}
$$

dimana

* $x_s(t)$ adalah posisi shock
* $f(u)$ adalah **flux function**
* $u_+$ dan $u_-$ adalah nilai solusi di kanan dan kiri shock

Kondisi ini memastikan bahwa **hukum konservasi tetap terpenuhi meskipun solusi tidak kontinu**.

### Rarefaction Wave

Jika kondisi awal dibalik

$$
u(x,0) =
\begin{cases}
3, & x < 0, \\
4, & x > 0
\end{cases}
$$

maka karakteristik menjadi

$$
x(t) =
\begin{cases}
3t + x_0, & x < 0, \\
4t + x_0, & x > 0
\end{cases}
$$

Karakteristik sekarang **menyebar**, sehingga terbentuk daerah tanpa karakteristik.

Daerah ini diisi oleh solusi yang berubah secara halus yang disebut **rarefaction wave**.


![Rarefaction](./images/characteristics_rarefaction.png)

### Self-Similar Solution

Untuk mencari solusi di daerah rarefaction, diasumsikan solusi berbentuk **self-similar**

$$
u(x, t) = u(\xi)
$$

dengan variabel

$$
\xi = x / t
$$

Substitusi ke persamaan Burgers

$$
\begin{aligned}
&u_t + u u_x = 0 \
&u' \frac{\partial \xi}{\partial t} + u u' \frac{\partial \xi}{\partial x} = 0 \
&u' \left ( -\frac{x}{t^2} + u \frac{1}{t} \right ) = 0 \
&u' \left ( -\xi + u\right ) = 0
\end{aligned}
$$

Persamaan terakhir memberikan dua kemungkinan solusi

$$
u' = 0
$$

atau

$$
u = \xi
$$

Solusi yang relevan pada daerah rarefaction adalah

$$
u = \xi
$$

yang memberikan profil solusi **linier antara dua keadaan awal**.

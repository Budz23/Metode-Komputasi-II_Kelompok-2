# Metode Elemen Hingga: sebuah pengantar

Metode Elemen Hingga (FEM) adalah salah satu cara untuk mengubah persamaan diferensial parsial (PDE) yang bersifat kontinu menjadi sistem persamaan diskret, baik linier maupun nonlinier. Metode ini sangat sering digunakan dalam bidang teknik dan mekanika padat, meskipun sebenarnya bisa diterapkan secara umum di berbagai bidang. FEM memiliki dasar teori yang kuat, terutama dari analisis fungsional, serta banyak sekali variasi penerapan dalam kasus-kasus nyata. Karena itu, pembahasan dan literatur tentang metode ini sangat luas. Catatan ini sendiri tidak membahas semuanya secara mendalam, melainkan hanya bertujuan untuk memberikan gambaran awal tentang konsep-konsep dasar dalam Metode Elemen Hingga dan bagaimana metode ini digunakan.


## Ruang Fungsi Dimensi Terbatas

Salah satu gagasan kunci dalam Elemen Hingga adalah bahwa kita akan mendekati fungsi kontinu $f(x)$ sebagai kombinasi linear dari sejumlah terbatas fungsi basis

$$
    f_h(x) = \sum_{i=0}^N w_i \phi_i(x)
$$


Penjelasan tiap komponen:

1. $f_h(x)$
   - Fungsi hampiran (aproksimasi) dari solusi sebenarnya.
   - Tanda "h" biasanya menunjukkan bahwa fungsi ini adalah hasil diskretisasi (mesh).

2. $Σ (sigma)$
   - Menyatakan penjumlahan dari i = 0 sampai N.
   - Artinya, semua kontribusi dari setiap elemen/basis dijumlahkan.

3. $N$
   - Jumlah total basis fungsi atau jumlah node dalam sistem.

4. $w_i$
   - Koefisien atau bobot (weight).
   - Nilai ini biasanya belum diketahui dan akan dicari dengan menyelesaikan sistem persamaan.

5. $φ_i(x)  (phi_i(x))$
   - Fungsi basis (basis function) atau fungsi bentuk (shape function).
   - Fungsi ini digunakan untuk membangun bentuk aproksimasi solusi.
   - Biasanya bernilai 1 di node ke-i dan 0 di node lainnya (untuk FEM sederhana).

Makna keseluruhan:
- Fungsi $f_h(x)$ dibentuk sebagai kombinasi linier dari fungsi-fungsi basis $φ_i(x)$,
   yang masing-masing dikalikan dengan bobot w_i.
- Dengan kata lain, solusi kontinu didekati menggunakan gabungan fungsi sederhana.

di mana $phi_i(x)$ adalah fungsi dasar dan $w_i$ adalah sekumpulan bobot. Kami telah melihat dekomposisi serupa untuk metode spektral. Namun alih-alih menggunakan fungsi dasar global seperti $sin(kx)$ atau polinomial Chebyshev, kita akan (sebagian besar) menggunakan polinomial sepotong-sepotong.


### Fungsi Linear Pecahan dalam 1-D (P1)

![](./images/P1_function_annotated_fig1.1LB.png)

Sebagai contoh, pertimbangkan fungsi "hubungkan titik" linier sepotong yang diilustrasikan di atas di mana kita telah menguraikan interval $I=[x_0,x_5]$ menjadi 5 sub-interval atau "elemen" $e_i = [x_{i},x_{i+1}]$ dengan $i=1ldots 4$. Selama masing-masing interval ini, fungsi $v_h$ adalah fungsi linier dari $x$ dan setiap segmen bersifat kontinu di tepi elemen. Gambar dimodifikasi dari Gambar 1.1 Larson dan Bengzon.

Sementara $v_h(x)$ adalah fungsi berkelanjutan yang dapat dievaluasi untuk nilai apa pun dari $x$, harus jelas bahwa sebenarnya hanya ada 6 derajat kebebasan yang sesuai dengan nilai pada simpul $vec v = [v_0,v_1,v_2,v_3,ldots,v_5]$. Hanya dengan mengubah 6 angka ini, kita dapat menghasilkan sejumlah tak terbatas dari fungsi linier sepotong pada triangulasi ini. Kita akan menyebut ruang dimensi terbatas fungsi ini $cal V_h$, yang sepenuhnya dijelaskan oleh mesh elemen, dan interpolan di atas setiap elemen

### Fungsi Dasar Global

Mengingat ruang fungsi diskrit $cal V_h$, kita harus dapat menggambarkan fungsi apa pun $v_hin cal V_h$ dalam hal sekumpulan fungsi dasar yang menjangkau ruang. Dalam kasus ruang fungsi linier 1-D sepotong-sepotong, fungsi dasar global diberikan oleh "Fungsi Topi"

$$
\phi_i(x) = \left\{
\begin{array}{ll}
(x - x_{i-1})/\Delta x_{i-1}, & x\in e_{i-1} \\
(x_i - x)/\Delta x_{i}, & x\in e_{i} \\
0, & \text{otherwise} \\
\end{array}
\right.
$$


1. $φ_i(x)$
   - Fungsi basis ke-i (shape function).
   - Digunakan untuk membangun solusi aproksimasi pada metode elemen hingga.

2. $x$$
   - Variabel posisi (misalnya koordinat dalam 1 dimensi).

3. $x_{i-1}, x_i$
   - Titik node (titik diskret pada domain).
   - $φ_i(x)$ bernilai maksimum di $x_i$.

4. $Δx_{i-1} = x_i - x_{i-1}$
   $Δx_i$    = $x_{i+1} - x_i$
   - Panjang elemen (jarak antar node).

5. $e_{i-1}$ dan $e_i$
   - Elemen (interval):
      $e_{i-1}$ = $[x_{i-1}, x_i]$
      $e_i$     = $[x_i, x_{i+1}]$

Makna tiap bagian:

- $(x - x_{i-1})$ / $Δx_{i-1}$, untuk $x ∈ e_{i-1}$
  - Fungsi naik secara linier dari 0 ke 1 saat $x$ bergerak dari $x_{i-1}$ ke $x_i$.

- $(x_i - x) / Δx_i$, untuk $x ∈ e_i$
  - Fungsi turun secara linier dari 1 ke 0 saat $x$ bergerak dari $x_i$ ke $x_{i+1}$.

- 0, selain itu
  - Fungsi hanya aktif di sekitar node i (bersifat lokal).

Sifat penting $φ_i(x)$:

1. $φ_i(x_i)$ = 1
2. $φ_i(x_j)$ = 0 untuk j ≠ i
3. Bentuknya segitiga (linear piecewise)
4. Hanya tidak nol pada dua elemen terdekat (lokal support)



Yang memiliki properti penting yaitu $phi_i(x_j) = delta_{ij}$. Fungsi topi khas ditunjukkan pada Gambar 1.2 dari Larson dan Bengson

![](./images/P1_Hat_function_fig1.2LB.png)

$$
f_h(x) = \sum_{i=0}^5 w_i\phi_i(x)
$$


### Fungsi Basis Lokal

Untuk membuat ini sedikit lebih jelas, ada baiknya melihat bagaimana fungsi dasar ini menginterpolasi $v$ pada satu elemen.

![](./images/P1LocalBases.png)


> **Gambar *3*: Definisi Fungsi Basis Global dan Lokal**

Pertimbangkan hanya bagian dari $v(x)$ yang terletak dalam elemen $e_i$. Jelas satu-satunya fungsi dasar global yang memiliki dukungan dalam elemen $e_i$ adalah $phi_i(x)$ dan $phi_{i+1}(x)$. Kami mengklaim bahwa di atas elemen ini

$$
v(x) = v_i\phi_i(x) + v_{i+1}\phi_{i+1}(x)\quad\text{for}\quad x\in [x_i,x_{i+1}]
$$


1. $v(x)$
   - Nilai fungsi aproksimasi pada titik x di dalam satu elemen.
   - Merupakan pendekatan solusi pada interval $[x_i, x_{i+1}]$.

2. $v_i dan v_{i+1}$
   - Nilai fungsi pada node ke-i dan node ke-(i+1).
   - Disebut juga derajat kebebasan (degrees of freedom).
   - Nilai ini biasanya diperoleh dari hasil penyelesaian sistem persamaan FEM.

3. $φ_i(x) dan φ_{i+1}(x)$
   - Fungsi basis (shape function) untuk node i dan i+1.
   - Mengatur kontribusi masing-masing node terhadap nilai di titik x.

4. $x ∈ [x_i, x_{i+1}]$
   - Menunjukkan bahwa rumus ini hanya berlaku dalam satu elemen (interval antar dua node).

Makna keseluruhan:

- Fungsi v(x) di dalam satu elemen dibentuk dari kombinasi linier
   dua fungsi basis yang terkait dengan dua node ujung elemen tersebut.

- Artinya, nilai di titik mana pun di antara $x_i$ dan $x_{i+1}$
   ditentukan oleh "campuran" nilai di kedua node tersebut.

Sifat penting:

1. $v(x_i) = v_i$
   - Karena $φ_i(x_i) = 1$ dan $φ_{i+1}(x_i)$ = 0$

2. $v(x_{i+1}) = v_{i+1}$
   - Karena $φ_i(x_{i+1}) = 0$ dan $φ_{i+1}(x_{i+1}) = 1$

3. Interpolasi linier
   - Di dalam elemen, fungsi berbentuk garis lurus (linear interpolation)


Ini jelas benar pada node karena properti bahwa $\phi_i(x_j) = \delta_{ij}$.


Di antara node, kami mengharapkan bahwa $\phi_i$ dan $\phi_{i+1}$ memberikan interpolasi linier. Jika kita hanya mempertimbangkan *fungsi dasar lokal* yang merupakan bagian dari $\phi_i$ dan $\phi_{i+1}$ dalam elemen $e_i$

\begin{align}
  N_0(x) & = (x_{i+1} - x)/(x_{i+1} - x_i)\\
  N_1(x) & = (x - x_i)/(x_{i+1} - x_i)\\
\end{align}

menjadi jelas bahwa ini hanyalah Polinomial Lagrange dari urutan 1 yang menginterpolasi fungsi linier apa pun selama interval $[x_i,x_{i+1}]$. Hanya untuk memeriksanya, perhatikan bahwa jika $x* = x_i + t \Delta x_i$ di mana $s$ adalah jarak pecahan melintasi elemen dengan panjang $\Delta x_i = x_{i+1} - x_i$, maka


\begin{align}
  N_0(x*) & = (x_{i+1} - x_i -s \Delta x_i)/\Delta x_i = (1 - s)\\
  N_1(x*) & = (x_i + s \Delta x_i - x_i)/\Delta x_i = s\\
\end{align}

so that within $e_i$

$$ 
   v(x) = v_i N_0(x) + v_{i+1}N_1(x) = v_i(1 - s)  + v_{i+1} s
$$
dengan $s\in[0,1]$

Jadi fungsi penuh kita $v(x)$ di atas domain penuh $\Omega = [x_0,x_5]$ hanyalah penyatuan potongan-potongan linier lokal ini. Sepanjang pengenalan ini kita akan sering berpindah antara tampilan global dan tampilan elemen tergantung pada kerangka acuan mana yang lebih nyaman.


### Ruang Fungsi Polinomial Orde Tinggi

Ruang fungsi linier terpotong sering disebut sebagai $P_1$, tetapi ada pilihan lain. Misalnya, ruang fungsi kuadratik terpotong ($P_2$) memiliki satu set fungsi basis lokal yang sederhana berupa Polinomial Lagrange orde 2 dan memerlukan tiga derajat kebebasan per simpul. Fungsi basis global dibuat dengan mudah dari fungsi lokal tersebut. Gambar menunjukkan subset fungsi basis untuk $P_2$ yang berbeda untuk derajat kebebasan di pusat dan tepi elemen (gambar asli dari [Introduction to Finite Element Methods]).(http://hplgit.github.io/INF5620/doc/pub/sphinx-fem)

## Interpolation of a function $f(x)$ onto $\cal V$

Mengingat serangkaian fungsi dasar $\phi_i(x)$ untuk ruang fungsi $\cal V$, harus jelas bahwa tidak semua fungsi dapat direpresentasikan dengan tepat, tanpa kesalahan dalam $\cal V$. Misalnya, mengingat fungsi halus $f(x)$, muncul pertanyaan bagaimana membangun perkiraan $f_h(x) \in \cal V$ yang meminimalkan beberapa ukuran kesalahan. Sebenarnya ada beberapa pendekatan. Perkiraan yang paling sederhana hanyalah *interpolasi* atau *proyeksi* dari $f$ ke $\cal V$, yang akan kita tunjukkan sebagai $pi f$, dan mendefinisikan sebagai fungsi $pi f \in \cal V$ sedemikian rupa sehingga


$$
    \pi f (x) = \sum_{i=0}^N f(x_i)\phi(x)
$$


1. $π f(x)$
   - Proyeksi atau aproksimasi dari fungsi f(x) ke dalam ruang fungsi diskret.
   - Dibaca: "pi dari f", yaitu pendekatan f menggunakan basis FEM.

2. $f(x)$
   - Fungsi asli (kontinu) yang ingin didekati.

3. $Σ (sigma)$
   - Penjumlahan dari $i = 0$ sampai $N$ (semua node).

4. $f(x_i)$
   - Nilai fungsi asli pada node ke-$i$.
   - Digunakan sebagai bobot dalam kombinasi linier.

5. $φ_i(x)$
   - Fungsi basis (shape function) ke-$i$.
   - Menentukan kontribusi nilai di node $i$ terhadap titik $x$$.

6. $N$
   - Jumlah total node atau basis fungsi.


Sifat penting:

1. $π f(x_i) = f(x_i)$
   - Aproksimasi tepat di titik node.

2. Fungsi bersifat piecewise (per elemen)
   - Di setiap elemen, bentuknya mengikuti fungsi basis (biasanya linier).

3. Bergantung pada mesh
   - Semakin banyak node ($N$ besar), biasanya aproksimasi semakin akurat.



Mengingat sifat interpolasi dari polinomial Lagrange, seharusnya jelas bahwa

$$
\pi f(x_j) = f(x_j)
$$

yaitu kedua fungsi memiliki kesalahan nol pada derajat kebebasan.

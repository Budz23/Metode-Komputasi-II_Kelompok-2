## Laplasian 9-Titik 

## Operator Laplasian

Operator Laplasian pada fungsi \( u(x, y) \) adalah:

$$\nabla^2 u = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2}$$

Dalam grid diskrit dengan jarak grid $( \Delta x = \Delta y = h)$,  
turunan kedua biasanya dihitung dengan metode **finite difference**.

Pendekatan lain untuk mendiskritisasi Laplasian $\nabla ^2 $ disebut 9-point stencil. Stencil ini didefinisikan sebagai
$$ \nabla ^2u_{ij}=\frac{1}{6\Delta x^2}\left( 4u_{i-1,j}+4u_{i+1,j}+4u_{i,j-1}+4u_{i,j+1}+u_{i-1,j-1}+u_{i+1,j-1}+u_{i+1,j+1}+u_{i-1,j+1}-20u_{i,j}\right)$$

Artinya kita menggunakan 9 titik grid

## Ekspansi Taylor di Sekitar $(x_i, y_j)$

Kita gunakan ekspansi Taylor di sekitar titik $(x_i, y_j)$.

**Titik kanan**  
$u_{i+1,j} = u + h u_x + \frac{h^2}{2} u_{xx} + \frac{h^3}{6} u_{xxx} + \frac{h^4}{24} u_{xxxx} + \ldots
$

**Titik kiri**  
$u_{i-1,j} = u - h u_x + \frac{h^2}{2} u_{xx} - \frac{h^3}{6} u_{xxx} + \frac{h^4}{24} u_{xxxx}$

Jika dijumlahkan:  
$u_{i+1,j} + u_{i-1,j} = 2u + h^2 u_{xx} + \frac{h^4}{12} u_{xxxx}$

## Skema Stencil Laplasian

Masukkan ke formula stencil:  

$4(u_{i+1,j} + u_{i-1,j} + u_{i,j+1} + u_{i,j-1})$ dan  $(u_{\text{diagonal}})$

Setelah disederhanakan diperoleh:  

$ \nabla^2 u = \frac{1}{6h^2} \left( 4(\text{cross}) + (\text{diagonal}) - 20u \right)$

Kita juga dapat menemukan LTE seperti sebelumnya, sehingga diperoleh
$$ \tau _{ij}=\frac{\Delta x^2}{12}\left( u_{xxxx}+2u_{xxyy}+u_{yyyy}\right) +\mathcal{O}(\Delta x^4).$$

## Analisis Error Skema Laplasian

Hasil ekspansi memberikan error:  

$tau_{ij} = \frac{h^2}{12} \left( u_{xxxx} + 2u_{xxyy} + u_{yyyy} \right) + O(h^4)$

Karena:  

$u_{xxxx} + 2u_{xxyy} + u_{yyyy} = \nabla^2(\nabla^2 u) = \nabla^4 u$

Maka:  

$ tau_{ij} = \frac{h^2}{12} \nabla^4 u $

Jadi apa yang kita peroleh? Jika kita melihat pada $\tau _{ij}$, kita sebenarnya dapat menuliskan suku turunannya sebagai
$$ u_{xxxx}+2u_{xxyy}+u_{yyyy}=\nabla ^2(\nabla ^2u)\equiv \nabla ^4u, $$
yaitu Laplasian dari Laplasian (disebut operator biharmonik). 
Jika kita kembali ke PDE asli, kita dapat langsung menemukan bentuk galat ini dengan mencatat bahwa
$$ \begin{aligned}\nabla ^2u&=f\\ \nabla ^2(\nabla ^2u)&=\nabla ^2f\end{aligned} $$
Dengan demikian, kita dapat menghitung suku galat secara langsung dengan mengambil Laplasian dari ruas kanan PDE asli!

Ini juga dapat menghasilkan metode orde lebih tinggi. Jika kita mendefinisikan
$$ f_{ij}=f(x_i,y_j)+\frac{\Delta x^2}{12}\nabla ^2f(x_i,y_j) $$
maka kita dapat menghilangkan suku berikutnya dari Local Truncation Error (LTE)! 

Kita bahkan bisa melakukan ini meskipun hanya memiliki fungsi $ f(x,y) $ pada sekumpulan titik diskret, karena kita dapat menggunakan  5-point stencil.

# **Review:  Finite Differences**
Finite Differences adalah ekspresi yang mengaproksimasi turunan dari suatu fungsi yang dievaluasi pada sekumpulan titik yang sering disebut stensil. Ekspresi ini dapat hadir dalam berbagai bentuk, termasuk jenis stensil, tingkat ketelitian, dan orde turunan. Dalam materi ini kita akan mengulas proses penurunan, analisis kesalahan, dan penerapan Finite Differences.

## **Penurunan Finite Differences (beda hingga)**

Pendekatan umum dalam menurunkan rumus Finite Differences seharusnya sudah dikenal, setidaknya untuk beda hingga orde pertama.
Terdapat tiga cara berbeda untuk mendefinisikan turunan pada suatu titik $x_i$

​
 .

$$
    u'(x_i) = \lim_{\Delta x \rightarrow 0} \left \{ \begin{aligned} 
        &\frac{u(x_i + \Delta x) - u(x_i)}{\Delta x} & \equiv D_+ u(x_i)\\
        &\frac{u(x_i + \Delta x) - u(x_i - \Delta_x)}{2 \Delta x} & \equiv D_0 u(x_i)\\
        &\frac{u(x_i) - u(x_i - \Delta_x)}{\Delta x} & \equiv D_- u(x_i).
    \end{aligned} \right .
$$

![Approximations to $u'(x)$](./images/fd_basic.png)

Jika alih-alih membiarkan $\Delta x \rightarrow 0$ , kita menyusun suatu hampiran (approximation) terhadap kemiringan 
$u'(x_i)$, maka definisi turunan yang kita miliki dapat langsung dipandang sebagai hampiran(approximation) terhadap turunan ketika $\Delta x$ mungkin kecil tetapi tidak sama dengan nol.

Untuk selanjutnya dalam ulasan ini, kita akan membahas pendekatan yang lebih sistematis dalam menurunkan hampiran (approximation) tersebut, serta menemukan hampiran dengan tingkat ketelitian yang lebih tinggi, hampiran (approximation) untuk turunan orde tinggi, dan memahami galat yang terkait dengan hampiran (approximation) tersebut.



## **Interpolating Polynomials (Polinomial Interpolasi)**

Salah satu cara untuk menurunkan hampiran beda hingga adalah dengan menentukan polinomial interpolasi yang melalui stensil yang diberikan, kemudian mendiferensiasikannya secara langsung. Diberikan $N+1$ titik $(x_0,u(x_0)), (x_1,u(x_1)), \ldots, (x_{N},u(x_{N}))$ dengan asumsi bahwa semua $x_i$ bernilai unik, maka polinomial interpolasi $P_N(x)$ dapat ditulis sebagai

$$
    P_N(x) = \sum^{N}_{i=0} u(x_i) \ell_i(x)
$$

where 

$$
    \ell_i(x) = \prod^{N}_{j=0, j \neq i} \frac{x - x_j}{x_i - x_j} = \frac{x - x_0}{x_i - x_0} \frac{x - x_1}{x_i - x_1} \cdots \frac{x - x_{i-1}}{x_i - x_{i-1}}\frac{x - x_{i+1}}{x_i - x_{i+1}} \cdots \frac{x - x_{N}}{x_i - x_{N}}
$$

Perhatikan bahwa $\ell_i(x_i) = 1$ dan untuk setiap $\forall j\neq i berlaku  ~~ \ell_i(x_j) = 0$. 

Karena kita telah mengetahui cara mendiferensiasikan suatu polinomial, seharusnya kita dapat menghitung hampiran beda hingga yang diberikan berdasarkan titik-titik data tersebut.

#### **Contoh: Stensil 2 Titik**
Misalkan kita memiliki dua titik untuk menyusun hampiran turunan. Polinomial interpolasi yang melalui kedua titik tersebut merupakan fungsi linear dengan bentuk
$$
    P_1(x) = u(x_0) \frac{x - x_1}{x_0 - x_1} + u(x_1) \frac{x - x_0}{x_1 - x_0}.
$$
Turunkan hampiran yang berpusat di $x_0$ dari polinomial ini.

Menurunkan $P_1(x)$ menghasilkan
$$
    P'_1(x) = u(x_0) \frac{1}{x_0 - x_1} + u(x_1) \frac{1}{x_1 - x_0}.
$$

Jika kita memungkinkan jarak antara $x_0$ dan $x_1$ menjadi $\Delta x = x_1 - x_0$, maka kita dapat menuliskannya sebagai

$$
    P'_1(x) = \frac{u(x_1) - u(x_0)}{\Delta x}
$$
yang merupakan bentuk umum dari $D_-u(x)$ dan $D_+u(x)$ di atas.

Jika kita perluas ini menjadi tiga titik, kita memperoleh polinomial interpolasi

$$
    P_2(x) = u(x_0) \frac{x - x_1}{x_0 - x_1} \frac{x - x_2}{x_0 - x_2} + u(x_1) \frac{x - x_0}{x_1 - x_0} \frac{x - x_2}{x_1 - x_2} + u(x_2) \frac{x - x_0}{x_2 - x_0} \frac{x - x_1}{x_2 - x_1}.
$$

Menurunkan persamaan ini menghasilkan
$$\begin{aligned}
    P'_2(x) &= u(x_0) \left( \frac{1}{x_0 - x_1} \frac{x - x_2}{x_0 - x_2} + \frac{x - x_1}{x_0 - x_1} \frac{1}{x_0 - x_2}\right )+ u(x_1) \left ( \frac{1}{x_1 - x_0} \frac{x - x_2}{x_1 - x_2} + \frac{x - x_0}{x_1 - x_0} \frac{1}{x_1 - x_2} \right )+ u(x_2)\left ( \frac{1}{x_2 - x_0} \frac{x - x_1}{x_2 - x_1} + \frac{x - x_0}{x_2 - x_0} \frac{1}{x_2 - x_1} \right ) \\
    &= u(x_0) \left(\frac{x - x_2}{2 \Delta x^2} + \frac{x - x_1}{2 \Delta x^2} \right )+ u(x_1) \left ( \frac{x - x_2}{-\Delta x^2} + \frac{x - x_0}{-\Delta x^2} \right )+ u(x_2)\left ( \frac{x - x_1}{2\Delta x^2} + \frac{x - x_0}{2 \Delta x^2} \right ) \\
    &=\frac{u(x_0)}{2\Delta x^2} (2x - x_2 - x_1)+ \frac{u(x_1)}{-\Delta x^2} ( 2x - x_2 - x_0)+ \frac{u(x_2)}{2\Delta x^2}( 2x - x_1 - x_0) \\
    &=\frac{u(x_0)}{2\Delta x^2} (2x - x_2 - x_1)+ \frac{u(x_1)}{-\Delta x^2} ( 2x - x_2 - x_0)+ \frac{u(x_2)}{2\Delta x^2}( 2x - x_1 - x_0).
\end{aligned}$$

Jika sekarang kita evaluasi turunan di $x_1$, dengan asumsi bahwa ini adalah titik pusatnya, kita akan peroleh
$$\begin{aligned}
    P'_2(x_1) &= \frac{u(x_0)}{2\Delta x^2} (x_1 - x_2)+ \frac{u(x_1)}{-\Delta x^2} ( x_1 - x_2 + x_1 - x_0)+ \frac{u(x_2)}{\Delta x^2}( x_1 - x_0) \\
    &= \frac{u(x_0)}{2\Delta x^2} (-\Delta x)+ \frac{u(x_1)}{-\Delta x^2} ( -\Delta x + \Delta x)+ \frac{u(x_2)}{\Delta x^2}( 2\Delta x) \\
    &= \frac{u(x_2) - u(x_0)}{2 \Delta x}
\end{aligned}$$
sehingga diperoleh hampiran ketiga yang telah disebutkan sebelumnya.


## **Taylor-Series Methods (Metode Deret Taylor)**
Cara lain untuk menurunkan hampiran beda hingga dapat dilakukan dengan menggunakan deret Taylor dan metode koefisien tak tentu.
$$u(x) = u(x_n) + (x - x_n) u'(x_n) + \frac{(x - x_n)^2}{2!} u''(x_n) + \frac{(x - x_n)^3}{3!} u'''(x_n) + \mathcal{O}((x - x_n)^4)$$

Misalkan kita ingin menurunkan hampiran turunan pertama dengan tingkat ketelitian orde dua, seperti yang baru saja kita lakukan. Hampiran ini memerlukan nilai-nilai $(x_{n+1}, u(x_{n+1}))$ dan $(x_{n-1}, u(x_{n-1}))$. Kita dapat menyatakan nilai-nilai ini melalui hampiran deret Taylor di atas sebagai

$$\begin{aligned}
    u(x_{n+1}) &= u(x_n) + (x_{n+1} - x_n) u'(x_n) + \frac{(x_{n+1} - x_n)^2}{2!} u''(x_n) + \frac{(x_{n+1} - x_n)^3}{3!} u'''(x_n) + \mathcal{O}((x_{n+1} - x_n)^4) \\
    &= u(x_n) + \Delta x u'(x_n) + \frac{\Delta x^2}{2!} u''(x_n) + \frac{\Delta x^3}{3!} u'''(x_n) + \mathcal{O}(\Delta x^4) 
\end{aligned}$$

dan
$$\begin{aligned}
    u(x_{n-1}) &= u(x_n) + (x_{n-1} - x_n) u'(x_n) + \frac{(x_{n-1} - x_n)^2}{2!} u''(x_n) + \frac{(x_{n-1} - x_n)^3}{3!} u'''(x_n) + \mathcal{O}((x_{n-1} - x_n)^4) \\
&= u(x_n) - \Delta x u'(x_n) + \frac{\Delta x^2}{2!} u''(x_n) - \frac{\Delta x^3}{3!} u'''(x_n) + \mathcal{O}(\Delta x^4) 
\end{aligned}$$

Untuk menentukan bagaimana menggabungkan kedua nilai ini menjadi suatu ekspresi turunan, kita asumsikan hampiran kita memiliki bentuk
$$u'(x_n) + R(x_n) = A u(x_{n+1}) + B u(x_n) + C u(x_{n-1})$$
dengan $R(x_n)$  merupakan galat dari hampiran ini.

Substitusi hampiran deret Taylor menghasilkan
$$u'(x_n) + R(x_n) = A \left ( u(x_n) + \Delta x u'(x_n) + \frac{\Delta x^2}{2!} u''(x_n) + \frac{\Delta x^3}{3!} u'''(x_n) + \mathcal{O}(\Delta x^4)\right ) + B u(x_n) + C \left ( u(x_n) - \Delta x u'(x_n) + \frac{\Delta x^2}{2!} u''(x_n) - \frac{\Delta x^3}{3!} u'''(x_n) + \mathcal{O}(\Delta x^4) \right )$$

Karena kita menginginkan $R(x_n) = \mathcal{O}(\Delta x^2)$, kita ingin semua suku yang berorde lebih rendah dari ini hilang, kecuali suku-suku yang mengalikan $u'(x_n)$, karena suku-suku tersebut harus berjumlah 1 untuk menghasilkan hampiran yang kita inginkan. Dengan mengelompokkan suku-suku yang memiliki turunan $u^{(k)}(x_n)$ yang sama, kita memperoleh serangkaian ekspresi untuk koefisien $A$, $B$, dan $C$ berdasarkan fakta bahwa kita menginginkan suatu hampiran terhadap $u'(x_n)$. Suku-suku untuk $n=0$ yang dikelompokkan adalah 
$A + B + C$dan ditetapkan sama dengan 0 karena kita menginginkan suku $u(x_n)$tersebut hilang.

$$\begin{aligned}
    u(x_n): & \quad A + B + C = 0 \\
    u'(x_n): & \quad A \Delta x - C \Delta x = 1  \\
    u''(x_n): & \quad A \frac{\Delta x^2}{2} + C \frac{\Delta x^2}{2} = 0 
\end{aligned}$$

Persamaan terakhir ini menghasilkan $\Rightarrow A = -C$. Dengan menggunakan hasil ini pada persamaan kedua, diperoleh $A = \frac{1}{2 \Delta x}$ dan $C = -\frac{1}{2 \Delta x}$. Persamaan pertama kemudian memberikan $B = 0$. Menggabungkan semuanya menghasilkan ekspresi kita sebelumnya, beserta dengan estimasi galatnya:

$$u'(x_n) + R(x_n) = \frac{u(x_{n+1}) - u(x_{n-1})}{2 \Delta x} + \frac{1}{2 \Delta x} \frac{\Delta x^3}{3!} u'''(x_n) + \mathcal{O}(\Delta x^4) + \frac{1}{2 \Delta x} \frac{\Delta x^3}{3!} u'''(x_n) + \mathcal{O}(\Delta x^4) $$

$$R(x_n) = \frac{\Delta x^2}{3!} u'''(x_n) + \mathcal{O}(\Delta x^3) = \mathcal{O}(\Delta x^2)$$

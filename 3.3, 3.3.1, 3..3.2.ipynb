## **3.3 Error Analysis (Analisis Kesalahan)** ##

Ketika kita menyelesaikan suatu persamaan secara numerik (misalnya menggunakan finite difference), hasil yang kita dapatkan biasanya bukan solusi yang benar-benar tepat, melainkan solusi pendekatan. 

Error didefinisikan sebagai selisih antara solusi aproksimasi dan solusi sebenarnya:
$$

E = U - \widehat{U}

$$

di mana:

- $U$ adalah vektor solusi pendekatan
- $\widehat{U}$ adalah vektor yang terdiri dari nilai solusi sebenarnya $u(x_i)$

sehingga

$$\widehat{U_i} = u(x_i)$$

artinya, setiap komponen dari vektor $\widehat{U}$ pada titik grid $x_i$.

Karena $E$ masih merupakan sebuah vektor, bagaimana perilaku norm dari $E$ untuk suatu nilai $Δx$ tertentu.
 
Untuk norm tak hingga (infinity norm), kita memiliki 

$$
||E||_{\infty} =
\max_{1 \le i \le m} |E_i|
=
\max_{1 \le i \le m} |U_i - u(x_i)|
$$

artinya, nilai tersebut adalah error terbesar di antara semua titik grid. Jika kita dapat menunjukkan bahwa

$$
||E||_{\infty} \rightarrow 0
\quad \text{ketika} \quad
\Delta x \rightarrow 0
$$

maka kita dapat menyatakan bahwa solusi aproksimasi $U_i$ pada setiap titik grid akan mendekati solusi sebenarnya, sehingga 

$$E_i \rightarrow 0 $$

Jika kita ingin menggunakan norm lain, biasanya kita mendefinisikan bentuk norm yang sedikit dimodifikasi yang juga melibatkan lebar grid $Δ_x$, yaitu

$$
||E||_1 =
\Delta x
\sum_{i=1}^{m}
|E_i|
$$

dan

$$
||E||_2 =
\left(
\Delta x
\sum_{i=1}^{m}
E_i^2
\right)^{1/2}
$$

Norm-norm tersebut disebut sebagai grid function norms, karena dihitung berdasarkan nilai fungsi pada titik-titik grid.

Error $E$ yang didefenisikan di atas dikenal sebagai global error. Salah satu tujuan utama dalam pembelajaran metode numerik adalah memahami bagaimana error $E$ berubah ketika dipengaruhi oleh faktor-faktor lain yang akan didefenisikan pada bagian selanjutnya.


### **3.3.1 Local Truncation Error (Kesalahan Pemotongan Lokal)** ###

Local Truncation Error dapat didefenisikan dengan cara mengganti solusi pendekatan $U_i$ dengan solusi sebenarnya $u(x_I)$. Karena persamaan aljabar yang kita gunakan hanyalah pendekatan dari masalah nilai batas (BVP) Boundary Value Problem yang asli, maka kita tidak mengharapkan bahwa solusi sebenarnya akan memenuhi persamaan tersebut secara tepat. 

Perbedaan yang muncul akibat hal tersebut disebut sebagai Local Truncation Error (LTE). Untuk pendekatan finite difference satu dimensi, kita memiliki persamaan:

$$
\frac{1}{\Delta x^2}(U_{i+1} - 2U_i + U_{i-1}) = f(x_i)
$$

Jika kita mengganti nilai pendekatan $U_i$ dengan solusi sebenarnya $u(x_i)$ pada persamaan tersebut, maka diperoleh:

$$
\tau_i = \frac{1}{\Delta x^2}\left(u(x_{i+1}) - 2u(x_i) + u(x_{i-1})\right) - f(x_i)
$$

Dalam bentuk ini, LTE masih balum terlalu informatif. Namun jika kita mengasumsikan bahwa fungsi $u(x_i)$ menggunakan deret Taylor, seperti yang dilakukan pada metode finite difference. 

Deret Taylor yang relevan adalah:

$$
u(x_{i\pm1}) = u(x_i) \pm u'(x_i)\Delta x
+ \frac{1}{2}u''(x_i)\Delta x^2
\pm \frac{1}{6}u'''(x_i)\Delta x^3
+ \frac{1}{24}u^{(4)}(x_i)\Delta x^4
+ O(\Delta x^5)
$$

Hal ini menghasilkan suatu ekspresi untuk $τ_i$ yaitu:
$$ \tau_i = \frac{1}{\Delta x^2} \left[ u''(x_i)\Delta x^2 + \frac{1}{12}u^{(4)}(x_i)\Delta x^4 + O(\Delta x^5) \right] - f(x_i) $$ $$ = u''(x_i) + \frac{1}{12}u^{(4)}(x_i)\Delta x^2 + O(\Delta x^4) - f(x_i) $$ $$ = \frac{1}{12}u^{(4)}(x_i)\Delta x^2 + O(\Delta x^4) $$

dimana kita mencatat bahwa soluhar $u''(x) = f(x)$

selama $ u^{(4)}(x_i) $ tetap terbatas(halus) kita tahu bahwa $\tau_i \rightarrow 0$ ketika $\Delta x \rightarrow 0$

Kita juga dapat menuliskan vektor LTE sebagai
$$
    \tau = A \widehat{U} - F
$$

yang mengimplikasikan
$$
    A\widehat{U} = F + \tau.
$$
Dapat diperhatikan bahwa disini yang digunakan adalah solusi eksak yang dievaluasi

### **3.3.2 Global Error(Kesalahan Global)** ###

Kesalahan global adalah selisi antara solusi numerik yang dihitung dengan metode numerik dan solusi eksak(solusi sebenarnya)pada suatu titik atau pada seluruh domain. Untuk menghubungkan kesalahan global dan LTE, kita dapat mensubtitusikan $E = U - \widehat{U}$ ke dalam ekspresi kita untuk LTE sehingga diperoleh $$ AE = -\tau. $$ ini berarti bahwa kesalahan global adalah solusi dari pendekatan dari persamaan yang telah kita definisikan untuk pendekatan tersebut, tetapi dengan $\tau$ sebagai fungsi pemaksa (fungsi pemaksaan) bukan $F$.

Hal ini juga menyiratkan bahwa kesalahan global $E$ dapat dianggap sebagai suatu pendekatan terhadap BVP yang serupa dengan yang kita mulai sebelumnya, yaitu $$ e''(x) = -\tau(x) \quad \Omega = [0, 1] \\ e(0) = 0 \quad e(1) = 0. $$

Kita dapat menyelesaikan ODE ini secara langsung dengan melakukan integrasi dua kali, sehingga untuk orde dua utama diperoleh $$ \begin{aligned} e(x) &\approx -\frac{1}{12} \Delta x^2 u''(x) + \frac{1}{12} \Delta x^2 (u''(0) + x (u''(1) - u''(0))) \\ &= \mathcal{O}(\Delta x^2) \rightarrow 0 \quad \text{ketika} \quad \Delta x \rightarrow 0. \end{aligned} $$

### Dimensional Splitting


1. KONSEP DASAR

Pada kasus difusi 2D, bentuk operator umumnya adalah:

$$
\frac{\partial U}{\partial t} = \frac{\partial^2 U}{\partial x^2}
+ \frac{\partial^2 U}{\partial y^2}
$$

Jika diselesaikan langsung (coupled), perhitungan menjadi cukup berat.
Karena itu digunakan metode splitting, yaitu memisahkan arah x dan y.

2. METODE SPLITTING DUA LANGKAH


Langkah 1 (arah x):

$$
U^*_{ij} = U^n_{ij} + \frac{\Delta t}{2}
\left( D^2_x U^n_{ij} + D^2_x U^*_{ij} \right)
$$

Langkah 2 (arah y):

$$
U^{n+1}_{ij} = U^*_{ij} + \frac{\Delta t}{2}
\left( D^2_y U^*_{ij} + D^2_y U^{n+1}_{ij} \right)
$$

Interpretasi:
- Tahap pertama: difusi hanya pada arah x
- Tahap kedua: difusi hanya pada arah y

3. APAKAH METODE INI VALID?


Metode ini akan mendekati solusi sebenarnya ketika:

$$
\Delta t \rightarrow 0
$$

Alasannya adalah karena operator turunan bersifat komutatif:

$$
\frac{\partial^2}{\partial x^2}
\left( \frac{\partial^2 U}{\partial y^2} \right)
=
\frac{\partial^2}{\partial y^2}
\left( \frac{\partial^2 U}{\partial x^2} \right)
$$

Artinya:
urutan pengerjaan arah x dan y tidak mempengaruhi hasil akhir
(dalam limit tertentu).

4. MASALAH BOUNDARY CONDITION


Karena ada variabel antara:

$$
U^*
$$

maka dibutuhkan boundary condition tambahan.

Permasalahannya:
- $U^*$ tidak berada tepat di waktu $t_n$ atau $t_{n+1}$
- sehingga nilai batasnya tidak langsung tersedia

Pendekatan umum:

$$
t^* = t_n + \frac{\Delta t}{2}
$$

Sehingga boundary condition untuk $U^*$ diambil pada waktu tengah ini.

5. METODE ADI (ALTERNATING DIRECTION IMPLICIT)

Untuk memperbaiki konsistensi, digunakan metode ADI:

Langkah 1:

$$
U^*_{ij} = U^n_{ij} + \frac{\Delta t}{2}
\left( D^2_y U^n_{ij} + D^2_x U^*_{ij} \right)
$$

Langkah 2:

$$
U^{n+1}_{ij} = U^*_{ij} + \frac{\Delta t}{2}
\left( D^2_x U^*_{ij} + D^2_y U^{n+1}_{ij} \right)
$$

Perbedaan utama:
- Kedua arah (x dan y) tetap muncul di setiap langkah
- Tetapi hanya satu arah yang implicit di tiap tahap

6. INTERPRETASI METODE ADI


Metode ini dapat dipandang sebagai prediksi solusi pada waktu:

$$
t_n + \frac{\Delta t}{2}
$$

Sehingga penggunaan boundary condition menjadi lebih konsisten
dibanding metode splitting biasa.

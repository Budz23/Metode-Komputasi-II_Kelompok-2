### Stabilitas

Kita telah menunjukkan bahwa analog kontinu dari $E$, yaitu $e(x)$, memang menuju nol ketika $\Delta x \rightarrow 0$. Namun bagaimana dengan $E$? 

Alih-alih menunjukkan sesuatu yang didasarkan pada $e(x)$, mari kita kembali melihat sistem persamaan awal untuk galat global:

$$
A_{\Delta x} E_{\Delta x} = - \tau_{\Delta x}
$$

di mana sekarang kita menyatakan suatu realisasi tertentu dari sistem tersebut dengan menggunakan jarak grid yang bersesuaian, yaitu $\Delta x$.

Jika kita dapat membalik (invert) $A_{\Delta x}$, maka kita dapat menghitung $E_{\Delta x}$ secara langsung. Dengan mengasumsikan bahwa hal tersebut dapat dilakukan dan dengan mengambil norma yang sesuai, kita peroleh

$$
\begin{aligned}
E_{\Delta x} &= (A_{\Delta x})^{-1} \tau_{\Delta x} \\
||E_{\Delta x}|| &= ||(A_{\Delta x})^{-1} \tau_{\Delta x}|| \\
& \leq ||(A_{\Delta x})^{-1} ||~|| \tau_{\Delta x}||
\end{aligned}
$$

Kita telah mengetahui bahwa $\tau_{\Delta x} \rightarrow 0$ ketika $\Delta x \rightarrow 0$ pada contoh kita. 

Jika kita dapat membatasi norma dari matriks $(A_{\Delta x})^{-1}$ dengan suatu konstanta $C$ untuk $\Delta x$ yang cukup kecil, maka kita dapat menuliskan batas untuk galat global sebagai berikut:

$$
||E_{\Delta x}|| \leq C ||\tau_{\Delta x}||
$$

Hal ini menunjukkan bahwa $E_{\Delta x} \rightarrow 0$ setidaknya secepat $\tau_{\Delta x} \rightarrow 0$.


Kita dapat menggeneralisasi pengamatan ini untuk semua masalah BVP linear dengan mengasumsikan bahwa kita memiliki pendekatan beda hingga (finite difference) terhadap suatu BVP linear dalam bentuk

$$
A_{\Delta x} U_{\Delta x} = F_{\Delta x},
$$

di mana $\Delta x$ adalah jarak grid (grid spacing).

Kita mengatakan bahwa pendekatan tersebut *stabil* jika $(A_{\Delta x})^{-1}$ ada untuk setiap $\Delta x < \Delta x_0$ dan terdapat suatu konstanta $C$ sehingga

$$
||(A_{\Delta x})^{-1}|| \leq C \quad \forall \Delta x < \Delta x_0.
$$

### Konsistensi

Salah satu konsep yang berkaitan dan penting dalam diskretisasi (membagi ruang atau waktu menjadi titik-titik grid) setiap persamaan diferensial parsial (PDE) adalah bahwa metode tersebut harus konsisten dengan persamaan yang sedang kita aproksimasi. Jika

$$
||\tau_{\Delta x}|| \rightarrow 0 \quad \text{ketika} \quad \Delta x \rightarrow 0
$$

maka kita mengatakan bahwa suatu aproksimasi *konsisten* dengan persamaan diferensial tersebut.

### Konvergensi

Sekarang kita sudah memiliki semua bagian yang diperlukan untuk membahas **galat global** $E$.  
Suatu metode dikatakan *konvergen* jika

$$
||E_{\Delta x}|| \rightarrow 0 \quad \text{ketika} \quad \Delta x \rightarrow 0.
$$

Artinya, ketika jarak grid $\Delta x$ dibuat semakin kecil, maka **galat (kesalahan (error) antara nilai yang sebenarnya dengan nilai yang dihitung atau diperkirakan.) global juga semakin mendekati nol**.

Jika suatu pendekatan memiliki dua sifat berikut:

- **Konsisten**  
  $$
  ||\tau_{\Delta x}|| \rightarrow 0 \quad \text{ketika} \quad \Delta x \rightarrow 0
  $$

- **Stabil**  
  $$
  ||E_{\Delta x}|| \leq C ||\tau_{\Delta x}||
  $$

maka pendekatan tersebut **pasti konvergen**.

Pembahasan ini awalnya diturunkan untuk kasus **Boundary Value Problem (BVP) linear**, tetapi pada praktiknya kriteria ini sering juga berlaku untuk berbagai **aproksimasi beda hingga (finite difference)** lainnya.

Pernyataan konvergensi ini juga sering diperkuat menjadi

$$
\mathcal{O}(\Delta x^p) \text{ LTE } + \text{ stabilitas } \Rightarrow \mathcal{O}(\Delta x^p) \text{ galat global}.
$$

Artinya, jika **local truncation error (LTE)** berorde $\mathcal{O}(\Delta x^p)$ dan metode tersebut stabil, maka **galat globalnya juga akan berorde $\mathcal{O}(\Delta x^p)$**.

Biasanya, bagian yang paling sulit dalam proses ini adalah membuktikan stabilitas.  
Pada bagian berikutnya kita akan melihat, untuk contoh sederhana yang kita gunakan, bagaimana cara membuktikan stabilitas menggunakan norma 2 (2-norm).

### Stabilitas dalam Norma 2

Jika kita mengingat kembali definisi **stabilitas**, maka kita perlu menunjukkan bahwa untuk matriks $A$ yang telah kita definisikan sebelumnya,

$$
(A_{\Delta x})^{-1}
$$

**ada (eksis)** dan memenuhi

$$
||(A_{\Delta x})^{-1}|| \leq C \quad \forall \Delta x < \Delta x_0
$$

untuk suatu konstanta $C$.

### Bagaimana kita tahu bahwa $(A_{\Delta x})^{-1}$ ada?

Kita dapat mengasumsikan bahwa matriks $A$ memang **dapat diinvers (invertible)**.  
Namun pertanyaan berikutnya adalah: **apakah kita bisa membatasi nilai norma dari invers tersebut?**

Untuk menjawabnya, ingat bahwa **norma 2 dari matriks simetris** sama dengan **jari-jari spektralnya (spectral radius)**, yaitu

$$
||A||_2 = \rho(A) = \max_{1\leq p \leq m} |\lambda_p|.
$$

Di sini:
- $\lambda_p$ adalah **nilai eigen (eigenvalue)** dari matriks $A$.

### Hubungan dengan invers matriks

Karena invers dari matriks simetris juga **tetap simetris**, maka **nilai eigen dari $A^{-1}$ adalah kebalikan dari nilai eigen $A$**.

Hal ini berarti

$$
||A^{-1}||_2 = \rho(A^{-1}) = 
\max_{1\leq p \leq m} \left| \frac{1}{\lambda_p} \right|
= 
\frac{1}{\max_{1\leq p \leq m} |\lambda_p|}.
$$

Jika **tidak ada nilai eigen $\lambda_p$ dari $A$ yang sama dengan nol** untuk $\Delta x$ yang cukup kecil, dan nilai-nilai eigen lainnya **tetap terbatas (finite)** ketika

$$
\Delta x \rightarrow 0,
$$

maka kita dapat menyimpulkan bahwa **aproksimasi yang digunakan bersifat stabil**.

Nilai eigen dari matriks $A$ yang telah dibahas sebelumnya dapat dituliskan sebagai

$$
\lambda_p = \frac{2}{\Delta x^2} (\cos(p \pi \Delta x) - 1)
$$

dengan **vektor eigen yang bersesuaian** $v^p$ diberikan oleh

$$
v^p_j = \sin(p \pi j \Delta x)
$$

di mana:

- $v^p_j$ adalah **komponen ke-$j$** dari vektor eigen $v^p$  
- $j = 1, \ldots, m$ menyatakan **indeks titik pada grid**.

#### Periksa bahwa pasangan berikut benar-benar merupakan pasangan eigen (eigenpairs) dari matriks $A$

$$
\lambda_p = \frac{2}{\Delta x^2} (\cos(p \pi \Delta x) - 1)
$$

$$
v^p_j = \sin(p \pi j \Delta x)
$$

Artinya, kita perlu memverifikasi bahwa **nilai eigen** $\lambda_p$ dan **vektor eigen** $v^p$ tersebut benar-benar memenuhi persamaan dasar eigen:

$$
A v^p = \lambda_p v^p
$$

Dengan kata lain, jika matriks $A$ dikalikan dengan vektor $v^p$, hasilnya harus sama dengan **nilai eigen $\lambda_p$ dikalikan dengan vektor $v^p$**.


$$
\begin{aligned}
(A v^p)_j &= \frac{1}{\Delta x^2} (v^p_{j-1} - 2 v^p_j + v^p_{j+1}) \\
&= \frac{1}{\Delta x^2} \big(\sin(p \pi (j-1) \Delta x) - 2 \sin(p \pi j \Delta x) + \sin(p \pi (j+1) \Delta x)\big) \\
&= \frac{1}{\Delta x^2} \big(\sin(p \pi j \Delta x)\cos(p \pi \Delta x) - 2\sin(p \pi j \Delta x) + \sin(p \pi j \Delta x)\cos(p \pi \Delta x)\big) \\
&= \lambda_p v^p_j
\end{aligned}
$$

Artinya, jika matriks $A$ dikalikan dengan vektor $v^p$, hasilnya sama dengan **nilai eigen $\lambda_p$ dikalikan dengan komponen vektor tersebut**.  

Hal ini membuktikan bahwa:

- $\lambda_p$ adalah **nilai eigen (eigenvalue)** dari matriks $A$  
- $v^p$ adalah **vektor eigen (eigenvector)** yang bersesuaian.

#### Menghitung Nilai Eigen Terkecil

Jika kita dapat menunjukkan bahwa **nilai-nilai eigen tidak berada dekat dengan nol**, maka kita tahu bahwa $||A||_2$ akan **terbatas (bounded)**.  
Pada kasus ini, nilai-nilai eigen bernilai **negatif**, sehingga kita perlu menunjukkan bahwa nilainya **selalu lebih kecil dari nol**.

Nilai eigen diberikan oleh

$$
\lambda_p = \frac{2}{\Delta x^2} (\cos(p \pi \Delta x) - 1)
$$

Untuk memahami bagaimana perilaku ekspresi ini terhadap $\Delta x$, kita dapat menggunakan **deret Taylor**.


Dari ekspansi deret Taylor diperoleh bahwa **nilai eigen terkecil** adalah

$$
\begin{aligned}
\lambda_1 &= \frac{2}{\Delta x^2} (\cos(p \pi \Delta x) - 1) \\
&= \frac{2}{\Delta x^2} \left(-\frac{1}{2} p^2 \pi^2 \Delta x^2 + \frac{1}{24} p^4 \pi^4 \Delta x^4 + \mathcal{O}(\Delta x^6)\right) \\
&= -p^2 \pi^2 + \mathcal{O}(\Delta x^2).
\end{aligned}
$$

Perhatikan bahwa hasil ini juga memberikan **batas galat (error bound)**, karena nilai eigen ini juga akan menghasilkan **nilai eigen terbesar dari matriks invers**.

Dengan demikian kita dapat menuliskan batas berikut

$$
||E^{\Delta x}||_2 \leq ||(A^{\Delta x})^{-1}||_2 \, ||\tau^{\Delta x}||_2
\approx
\frac{1}{\pi^2} ||\tau^{\Delta x}||_2.
$$

Artinya, **galat global** dibatasi oleh konstanta $\frac{1}{\pi^2}$ dikalikan dengan **galat pemotongan (truncation error)**.
